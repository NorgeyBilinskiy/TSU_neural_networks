## <h1 id="aa" style="color:white;background:#B00000;padding:8px;border-radius:8px;font-family:newtimeroman;"> Задача №1: вариант №4 </h1>

0) Заданы синаптические коэффициенты:
$$
v_1 = v_2 = w_{11} = w_{12} = 1;  w_{21} = w_{22} = -1 ;
$$
Смещения всех нейронов равны нулю.

На вход подается значение $$x = 1,$$ желаемое значение выходов $$sigma_1 = sigma_2= 1.$$

1) Сначала вычислим значения, получаемые на скрытых нейронах:

$$
y_{11} = x \cdot v_1
$$

$$
y_{12} = x \cdot v_2
$$

Подставим значения v1 = 1, v2 = 1, x = 1. Так как входные значения больше 0, согласно функции активации они остаются без изменений

$$
y_{11} = 1 \cdot 1 = 1
$$

$$
y_{12} = 1 \cdot 1 = 1
$$

2) Теперь вычислим значения, получаемые на выходных нейронах:

$$
y_{21} = y_{11} \cdot w_{11} + y_{12} \cdot w_{12}
$$

$$
y_{22} = y_{12} \cdot w_{22} + y_{11} \cdot w_{21}$$

Подставим значения $$y_{11} = y_{12} = 1, w_{11} = w_{12} = 1;  w_{21} = w_{22} = -1$$ Так как функция активации данных нейронов линейная, значения остаются без изменений.

$$
y_{21} = 1 \cdot 1 + 1 \cdot 1 = 2
$$

$$
y_{22} = 1 \cdot (-1) + 1 \cdot (-1)= -2
$$

3) Вычисление функции потерь L

Функция потерь задается как:

$$
L = \frac{1}{2} \cdot [(\sigma_1 - y_1)^2 + (\sigma_2 - y_2)^2]
$$

Подставляем желаемое значение выходов $$sigma_1 = sigma_2 = 1.$$

$$
L = \frac{1}{2} \cdot [(1 - 2)^2 + (1 - (-2))^2] = \frac{1}{2} \cdot [1 + 9] = 5
$$

4) Вычисление градиента функции потерь для выходных нейронов

Градиент функции потерь для выходных нейронов вычисляется как:

$$
\frac{\partial L}{\partial y} = \sigma - y
$$

Таким образом:

$$
\frac{\partial L}{\partial y_{21}} = \sigma_1 - y_{21} = 1 - 2 = -1
$$

$$
\frac{\partial L}{\partial y_{22}} = \sigma_1 - y_{22} = 1 - (-2) = 3
$$

5) Вычисление градиента для нейронов 1-ого слоя

Градиент функции потерь для y11 будет вычисляться по формуле:
$$
\frac{\partial L}{\partial y_{11}} = \frac{\partial L}{\partial y_{21}} \cdot
 w_{11} + \frac{\partial L}{\partial y_{22}} \cdot w_{21}
$$

И примет значение:
$$
\frac{\partial L}{\partial y_{11}} = -1 \cdot 1 + 3 \cdot (-1) = -1 -3 = -4
$$

Гградиент функции потерь для y12 будет вычисляться по формуле:
$$
\frac{\partial L}{\partial y_{12}} = \frac{\partial L}{\partial y_{21}} \cdot
 w_{12} + \frac{\partial L}{\partial y_{22}} \cdot w_{22}
$$

И примет значение:
$$
\frac{\partial L}{\partial y_{12}} = -1 \cdot 1 + 3 \cdot (-1) = -1 -3 = -4
$$ \

Функция активации нейронов 1-ого слоя:

$$f(h)=h:  ℎ≥0$$
$$𝑓(ℎ)=−ℎ: ℎ<0$$

Так как в нашем случае на вход подавались положительные знаечния, и функция активации выдавала их без изменений, то градиент функции активации равен единице, что не изменяет уже полученные градиенты для нейронов первого слоя.

6) Вычисление градиента для веса u1.

Градиент функции потерь для y11 будет вычисляться по формуле:
$$
\frac{\partial L}{\partial u_{1}} = \frac{\partial L}{\partial y_{11}} \cdot
\frac{\partial y_{11}}{\partial u_{1}}
$$

Так как $$y_{11} = x \cdot u_1$$

То
$$
\frac{\partial y_{11}}{\partial u_{1}} = x
$$

Подставим значения:
$$
\frac{\partial L}{\partial u_{1}} = -4 \cdot 1 = -4
$$

7) Для обновления значения веса u1 используется правило градиентного спуска:

$$
v_1^{\text{новое}} = v_1^{\text{старое}} - \alpha \cdot \frac{\partial L}{\partial v_1}
$$

Подставим значения:
$$
v_1^{\text{новое}} = 1 - 0.1 \cdot (-4) = 1.4
$$



````python
# 0 Условия
v_1 = 1
v_2 = 1
w_11 = 1
w_12 = 1
w_21 = -1
w_22 = -1
x = 1
sigma_1 = 1
sigma_2 = 1
learning_rate = 0.1

# 1. Вычисляем значения на скрытых нейронах
y_11 = x * v_1
y_12 = x * v_2

# Так как значения положительные, после функции активации они остаются неизменными
y_11 = y_11
y_12 = y_12

# 2. Вычисляем значения на выходных нейронах
y_21 = y_11 * w_11 + y_12 * w_12
y_22 = y_11 * w_21 + y_12 * w_22

# 3. Вычисляем функцию потерь L
L = 0.5 * ((sigma_1 - y_21)**2 + (sigma_2 - y_22)**2)

# 4. Вычисляем градиенты для выходных нейронов
dL_dy21 = sigma_1 - y_21  # Градиент по y_21
dL_dy22 = sigma_2 - y_22  # Градиент по y_22

# 5. Вычисляем градиенты для нейронов первого слоя
dL_dy11 = dL_dy21 * w_11 + dL_dy22 * w_21
dL_dy12 = dL_dy21 * w_12 + dL_dy22 * w_22

# Так как входы на нейроны были положительные, градиенты не меняются
dL_dy11 = dL_dy11 if y_11 > 0 else -y_11
dL_dy12 = dL_dy12 if y_12 > 0 else -y_11

# 6. Вычисляем градиент для веса v_1
dL_dv1 = dL_dy11 * x

# 7. Обновляем вес с помощью правила градиентного спуска
v_1_new = v_1 - learning_rate * dL_dv1
print(f"Новое значение v_1: {v_1_new}")
````

0) Заданы синаптические коэффициенты:
$$
v_1 = v_2 = 1;  w_{1} = -1; w_{2} = 0.5 ;
$$
Смещения всех нейронов равны нулю.

На вход подается значение $$x = 0.5.$$

1) Вычисление активаций нейронов

1. Для первого нейрона:

$$
h_{11} = x \cdot 1 = 0.5 \cdot 1 = 0.5
$$

2. Для второго нейрона:

$$
h_{12} = x \cdot 1 = 0.5 \cdot 1 = 0.5
$$

3. Для выходного нейрона:

$$
h_{22} = h_{11} \cdot (-1) + h_{12} \cdot 0.5 = 0.5 \cdot (-1) + 0.5 \cdot 0.5 = -0.5 + 0.25 = -0.25
$$

2) Вероятности выхода нейронов

Теперь применим функцию активации p(h) для каждого нейрона, которая:
* Возвращает 0, если h <= -1;
* Возвращает 0.5(h+1), если -1 < h <= 1;
* Возвращает 1, если h > 1.

Выход каждого стохастического нейрона - это дискретная случайная велична, описываемая распределением:
$$
P[Y=0]=1-p(h)
$$
$$
P[Y=1]=p(h)
$$


1. Для первого нейрона h_1 = 0.5, следовательно применим 0.5(h+1):

$$
p(h_{11}) = 0.5 \cdot (0.5 + 1) = 0.5 ⋅ 1.5 = 0.75
$$
$$
P[Y=0]=1-p(h_{11}) = 1 - 0.75 = 0.25
$$
$$
P[Y=1]=p(h_{11}) = 0.75
$$


2. Для второго нейрона h_1 = 0.5, следовательно применим 0.5(h+1):

$$
p(h_{12}) = 0.5 \cdot (0.5 + 1) = 0.5 ⋅ 1.5 = 0.75
$$
$$
P[Y=0]=1-p(h_{12}) = 1 - 0.75 = 0.25
$$
$$
P[Y=1]=p(h_{12}) = 0.75
$$

3. Для выходного нейрона h_3 = -0.25, следовательно применим 0.5(h+1):

$$
p(h_{22}) = 0.5 \cdot (-0.25 + 1) = 0.5 ⋅ 0.75 = 0.375
$$
$$
P[Y=0]=1-p(h_{22}) = 1 - 0.375 = 0.625
$$
$$
P[Y=1]=p(h_{22}) = 0.375
$$


3) Математическое ожидание выхода сети

Теперь найдём математическое ожидание выхода сети. Для этого воспользуемся формулой:

$$
\mathbb{E}[Y] = 2p(h)-1
$$

1. Для первого нейрона:

$$
\mathbb{E}[Y_11] = 2*0.75-1 = 1.5 - 1 = 0.5
$$

2. Для второго нейрона:

$$
\mathbb{E}[Y_12] = 2*0.75-1 = 1.5 - 1 = 0.5
$$

3. Для выходного нейрона:

$$
\mathbb{E}[Y_22] = 2*0.375-1 = 0.25 - 1 = -0.25
$$

**Ответ: Математическое ожидание выхода сети = -0.25.**


````python
# Заданные параметры
x = 0.5
v1 = v2 = 1
w1 = -1
w2 = 0.5

# Функция для вычисления активации нейронов
def activation(h):
    if h <= -1:
        return 0
    elif -1 < h <= 1:
        return 0.5 * (h + 1)
    else:
        return 1

# Вычисление активаций нейронов
h11 = x * v1
h12 = x * v2
h22 = h11 * w1 + h12 * w2

# Применение функции активации для каждого нейрона
p_h11 = activation(h11)
p_h12 = activation(h12)
p_h22 = activation(h22)

# Вывод вероятностей выхода
P_Y0_h11 = 1 - p_h11
P_Y1_h11 = p_h11

P_Y0_h12 = 1 - p_h12
P_Y1_h12 = p_h12

P_Y0_h22 = 1 - p_h22
P_Y1_h22 = p_h22

# Вычисление математического ожидания выхода сети
E_Y11 = 2 * p_h11 - 1
E_Y12 = 2 * p_h12 - 1
E_Y22 = 2 * p_h22 - 1
print(f"Математическое ожидание выхода сети: {E_Y22}")
````

v_{12} = v_{21} = w_1 = w_2 = 1;

w_3 = 3

На вход подается значения
$$x_1 = 2$$
$$x_2 = 1$$
Желаемое значение выхода $$sigma = 3.$$

Функция активации нейронов y11, y12:
$$
h, если h => 0;
$$
$$
0, если h < 0.
$$

Функция активации нейрона y22 - линейная.

1) Сначала вычислим значения, получаемые на скрытых нейронах:

$$
h_{11} = x_1 \cdot v_{11} + x_2 \cdot v_{12} = 2 \cdot 2 + 1 \cdot 1 = 4 + 1 = 5
$$
$$
y_{11} = h_{11} = 5, т.к. h_{11}=>0
$$

$$
h_{12} = x_1 \cdot v_{21} + x_2 \cdot v_{22} + y_{11} \cdot w_3 = 2 \cdot 1 + 1 \cdot 2 + 5 \cdot 3 = 2 + 2 + 15 = 19
$$
$$
y_{12} = h_{12} = 19, т.к. h_{11}=>0
$$

2) Теперь вычислим значение, полученное на выходном нейроне:

$$
y_{22} = y_{11} \cdot w_{1} + y_{12} \cdot w_{2} = 5 \cdot 1 + 19 \cdot 1 = 5+19 = 24
$$
$$
y_{22} = h_{22} = 24$$ т.к. функция активации выходного нейрона линейная.

3) Вычисление функции потерь L

Функция потерь задается как:

$$
L(v_1) = \frac{1}{2} \cdot (\sigma - y_{22})^2
$$

Подставляем желаемое значение выхода sigma=3:

$$
L(v_1) = \frac{1}{2} \cdot (3 - 24)^2 = \frac{1}{2} \cdot 441 = 220.5
$$

4) Вычисление градиента функции потерь для выходного нейрона y22

Градиент функции потерь по y22 вычисляется как:

$$
\frac{\partial L}{\partial y22} = -(σ - y_{22})
$$

Подставляем sigma = 3 и y22 = 24

$$
\frac{\partial L}{\partial y22} = -(3 - 24) = - (-21) = 21
$$

5) Вычисление градиентов для весов, входящих в нейрон y22

Градиенты весов, входящих в выходной нейрон y22, рассчитываются с использованием правила цепного дифференцирования:

Для весов w1 и w2 имеем:

$$
\frac{\partial L}{\partial w_1} = \frac{\partial L}{\partial y_{22}} \cdot \frac{\partial y_{22}}{\partial w_1} = 21 \cdot y_{11} = 21 \cdot 5 = 105
$$
$$
\frac{\partial L}{\partial w_2} = \frac{\partial L}{\partial y_{22}} \cdot \frac{\partial y_{22}}{\partial w_2} = 21 \cdot y_{12} = 21 \cdot 19 = 399
$$

6) Вычисление градиентов для скрытого нейрона y12

Для скрытого нейрона y12 рассчитаем градиент функции потерь по его активации:

$$
\frac{\partial L}{\partial y_{12}} = \frac{\partial L}{\partial y_{22}} \cdot \frac{\partial y_{22}}{\partial y_{12}} = 21 \cdot w_2 = 21 \cdot 1 = 21
$$

Градиенты для весов, входящих в нейрон y12:

$$
\frac{\partial L}{\partial w_3} = \frac{\partial L}{\partial y_{12}} \cdot \frac{\partial y_{12}}{\partial w_3} = 21 \cdot y_{11} = 21 \cdot 5 = 105
$$
$$
\frac{\partial L}{\partial v_{21}} = \frac{\partial L}{\partial y_{12}} \cdot \frac{\partial y_{12}}{\partial v_{21}} = 21 \cdot x_1 = 21 \cdot 2 = 42
$$
$$
\frac{\partial L}{\partial v_{22}} = \frac{\partial L}{\partial y_{12}} \cdot \frac{\partial y_{12}}{\partial v_{22}} = 21 \cdot x_2 = 21 \cdot 1 = 21
$$

7) Обновление для скрытого нейрона y11

Для скрытого нейрона y11 градиент будет:

$$
\frac{\partial L}{\partial y_{11}} = \frac{\partial L}{\partial y_{22}} \cdot \frac{\partial y_{22}}{\partial y_{11}} + \frac{\partial L}{\partial y_{12}} \cdot \frac{\partial y_{12}}{\partial y_{11}} = 21 \cdot w_1 + 21 \cdot w_3 = 21 \cdot 1 + 21 \cdot 3 = 21 + 63 = 84
$$

Градиенты для весов, входящих в нейрон y11:

$$
\frac{\partial L}{\partial v_{11}} = \frac{\partial L}{\partial y_{11}} \cdot \frac{\partial y_{11}}{\partial v_{11}} = 84 \cdot x_1 = 84 \cdot 2 = 168
$$
$$
\frac{\partial L}{\partial v_{12}} = \frac{\partial L}{\partial y_{11}} \cdot \frac{\partial y_{11}}{\partial v_{12}} = 84 \cdot x_2 = 84 \cdot 1 = 84
$$

8) Обновим вес v12 с использованием градиентного спуска. Cкорость обучения равна $$ \eta = 0.02 $$

$$
v_{12}^{\text{new}} = v_{12} - \eta \cdot \frac{\partial L}{\partial v_{12}} = 1 - 0.02 \cdot 84 = 1 - 1.68 = -0.68
$$

**Ответ: Обновлённый вес v12 после 1-ой итерации -0.68**

````python
# Исходные данные
x1 = 2
x2 = 1
sigma = 3

# Синаптические коэффициенты
v11, v12 = 2, 1
v21, v22 = 1, 2
w1, w2, w3 = 1, 1, 3

# Шаг обучения
alpha = 0.02

# Шаг 1: Вычисление скрытых нейронов
h11 = x1 * v11 + x2 * v12
y11 = h11 if h11 >= 0 else 0

h12 = x1 * v21 + x2 * v22 + y11 * w3
y12 = h12 if h12 >= 0 else 0

# Шаг 2: Вычисление выхода нейрона y22
y22 = y11 * w1 + y12 * w2

# Шаг 3: Вычисление функции потерь
L = 0.5 * (sigma - y22) ** 2

# Шаг 4: Градиент функции потерь по y22
dL_dy22 = -(sigma - y22)

# Шаг 5: Градиенты для весов w1 и w2
dL_dw1 = dL_dy22 * y11
dL_dw2 = dL_dy22 * y12

# Шаг 6: Градиент для нейрона y12 и его весов
dL_dy12 = dL_dy22 * w2
dL_dw3 = dL_dy12 * y11
dL_dv21 = dL_dy12 * x1
dL_dv22 = dL_dy12 * x2

# Шаг 7: Градиент для нейрона y11 и его весов
dL_dy11 = dL_dy22 * w1 + dL_dy12 * w3
dL_dv11 = dL_dy11 * x1
dL_dv12 = dL_dy11 * x2

# Шаг 8: Обновление веса v12
v12_new = v12 - alpha * dL_dv12
print(f"Обновленный вес v12: {v12_new:.2f}")
````
